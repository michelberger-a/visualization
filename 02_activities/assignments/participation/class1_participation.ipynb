```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
# make random sample data here
np.random.seed(613)
x = np.arange(50) # says the range is 0 to 50, so its giving integers from 0 to 49
y = np.random.randint(0,100,50)

# start plot
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# bar chart instead of scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y) # same as before, but now changed to bars
# although not appropriate for the type of data

# and try below for horizontal bars
# ax.barh(x,y)
```

```python
# line plot in the same manner
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# histogram
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
fig, ax2 = plt.subplots(figsize=(5,3))
ax2.hist(x)
```

```python
# adding labels
# create graph
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

# add the text and titles
ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years since start')
```

```python
# modify labels and titles
# using dictionaries {} key value pairs

font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

# set labels with dicts
ax.set_title('Total growth over time', fontdict=font1, loc='left')
ax.set_ylabel('Total Growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
# modify data points to be different
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
           marker = '>', # can try o, +, >, *
           color='indigo') # changes colour of the points
```

```python
# modify data points to be different - connect with lines
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*', # can try o, +, >, *
           color='indigo', # changes colour of the points
           linestyle = '--', # connects with lines
           linewidth = 2) # width of the line
```

```python
# marker colours edge and fill and sizes
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*', # can try o, +, >, *
           color='indigo', # changes colour of the points
           linestyle = '--', # connects with lines
           linewidth = 2, # width of the line
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')
```

```python
# adding gridlines
# specify horizontal with y, vertical with x
# add ith ax.grid

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*', # can try o, +, >, *
           color='indigo', # changes colour of the points
           linestyle = '--', # connects with lines
           linewidth = 2, # width of the line
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax.grid(axis='y') # try with axis = 'x', axis = 'y' and neither (uses both)
```

```python
# modify into anything we want - play with what ever we want
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*', # can try o, +, >, *
           color='lightpink', # changes colour of the points
           linestyle = 'dashdot', # connects with lines
           linewidth = 2, # width of the line
           markeredgecolor = 'cyan',
           markerfacecolor = '#000000')

ax.grid(color='olive', linestyle=':')
```

```python
# create your own visual
# Data
barWidth = 0.25
bars1 = [12, 30, 1, 8, 22]
bars2 = [28, 6, 16, 5, 10]
bars3 = [29, 3, 24, 25, 17]

# Bar positions
r = np.arange(len(bars1))
r2 = r + barWidth
r3 = r2 + barWidth

# Plotting
fig, ax = plt.subplots(dpi=300)
ax.bar(r, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='var1')
ax.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='var2')
ax.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='var3')

# Xticks
ax.set_xlabel('group', fontweight='bold')
ax.set_xticks(r + barWidth)
ax.set_xticklabels(['A', 'B', 'C', 'D', 'E'])

# colours of background and colour themes
ax.set_facecolor('gainsboro')

# grid lines
ax.grid(axis='x')

# Legend and show
ax.legend()
plt.show()
```
