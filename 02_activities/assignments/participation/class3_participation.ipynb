```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
# creating our data

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50) # 50 random integers from 0 to 99
y2 = np.random.randint(0,100,50)
```

```python
# create a lineplot that shares two lines on the same set of axes
fig, ax = plt.subplots(figsize=(5, 3))

# plots both lines on the same axes
ax.plot(x, y1)
#ax.scatter(x, y1) # will add points to all lines for blue
ax.plot(x, y2) # key to use the same ax object

ax.set_xlabel('Months')
ax.set_ylabel('Score')
```

```python
# Adding a legend

fig, ax = plt.subplots(figsize=(5, 3))

# plots both lines on the same axes
ax.plot(x, y1, label = 'Person 1') # label is for the plot
ax.plot(x, y2, label = 'Person 1') # label is for the plot
ax.legend(loc="lower right") # will add the legend to the area specified

ax.set_xlabel('Months')
ax.set_ylabel('Score')
```

```python
# modify the legend

fig, ax = plt.subplots(figsize=(5, 3))

# plots both lines on the same axes
ax.plot(x, y1, label = 'Person 1') # label is for the plot legend
ax.plot(x, y2, label = 'Person 1') # label is for the plot legend
ax.legend(loc="lower right",
        frameon=True,
        fontsize=12,
        ncol=2,
        shadow=True) # see changes and how it affect the legend 
# we have the show, larger text and its now in one row because we have two cols

ax.set_xlabel('Months')
ax.set_ylabel('Score')
```

```python
# move legend outside of plot but keep in the container

fig, ax = plt.subplots(figsize=(5, 3))

# plots both lines on the same axes
ax.plot(x, y1, label = 'Person 1') # label is for the plot legend
ax.plot(x, y2, label = 'Person 1') # label is for the plot legend
ax.legend(loc="upper left",
          bbox_to_anchor = (1, 0.5)) # see changes and how it affect the legend 
# the legend is anchored now outside of the visual, but anchored at the top left of the legend box
# try using lower left, upper right, lower right etc
# also try different coordinates i.e. (1, 0.5) - as this is the location
# the units represent a score out of 100, basically 1 is 100% of figure size
# so (1, 0.5) means 100% across, but 50% down from the top

ax.set_xlabel('Months')
ax.set_ylabel('Score')
```

```python
# Text and Annotations

# use code for reproducibility

fig, ax = plt.subplots(figsize=(5, 3))

# plots both lines on the same axes
ax.scatter(x, y1, label = 'Person 1') 
ax.scatter(x, y2, label = 'Person 1') 
ax.legend(loc="lower right",
          bbox_to_anchor = (1,1))
ax.text(10,95, 'This value is important!') # those numbers are coordinates on the axes
# it represents a point at x=10, y=95 (even if an individual data point does not exist)

ax.set_xlabel('Months')
ax.set_ylabel('Score')
```

```python
# change the text at that data point
fig, ax = plt.subplots(figsize=(5, 3))

# plots both lines on the same axes
ax.scatter(x, y1, label = 'Person 1') 
ax.scatter(x, y2, label = 'Person 1') 
ax.legend(loc="lower right",
          bbox_to_anchor = (1,1))
ax.text(10,95, 'This value is important!',
        ha = 'center',
        color = 'red',
        size = 20) # those numbers are coordinates on the axes
# we have now changed the text to be large, red and centred about the coordinate

ax.set_xlabel('Months')
ax.set_ylabel('Score')
```

```python
# Transforms
# position text annotations

# specify the coordinates if they are from the data, the axes or the figure overall
```

```python
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5, ". Data: (1,5)", transform=ax.transData) # uses data point
# would expect to find it at a data point with coords 1, 5
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes) # uses the axes object
# would expect to find it halfway across and one tenth up from bottom according to the visual
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform = fig.transFigure) # uses the figure object
# would expect to find it two tenths across and two tenths up from bottom according to the overall figure container

# the . Axes/Data - tell us what we are referencing for each
```

```python
# annotate with text and arrows
# does not need to be only text

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc='lower right')

ax.annotate('This is important!', 
            xy = (10,95), # this is where the tip of the arrow points
            xytext=(20,94), # this is where the text will print, as well as the flat end of the arrow
            arrowprops=dict(facecolor = 'black', shrink = 0.05)) # colour of the arrow
```

```python
# modify the arrow
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc='lower right')

ax.annotate('This is important!', 
            xy = (10,95),
            xytext=(20,94), 
            arrowprops=dict(facecolor = 'hotpink', arrowstyle = 'wedge')) #colour of the arrow 
# other types of arrows edge, <->, }, check the documentation
```

##### Axis labels (x,y) #####

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label='Person 1')
ax.scatter(x, y2, label='Person 2')
ax.legend(loc = 'lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) # to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) # remove only labels
```

```python
# major locator
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # tells us we want only 3 tick marks on x
# change the number in () to see how many ticks you get
```

```python
# limiting tick mark intervals, a particular interval instead

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')
#ax.xaxis.set_major_locator(plt.MaxNLocator(3))
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # tick mark every 5 units
# common and highly useful for creating logical/visually appealing intervals
```

```python
# rotate axis labels
# use xticks() to modify the labels established on the axis
# changes alignment and rotation

# try accessing font and font size
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

plt.xticks(rotation=45, 
           ha='right', 
           fontsize=20, 
           font='serif') # fontsize is a standard call
plt.yticks(rotation = 50,
           ha='left', 
           fontsize=15,
           font='serif')
```

```python
# modify axis title format and font
# define font dictionary
font1 = {'family':'serif', 'color':'purple'}

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'Person 1')
ax.scatter(x, y2, label = 'Person 2')
ax.legend(loc = 'lower right')

# change x ticks 
plt.xticks(rotation=45, 
           ha='right', 
           fontsize=20, 
           font='serif')

# change x-axis title - now labelled to what we want
plt.xlabel('Shiny New X Axis', fontsize=18, fontdict=font1)
```

```python
# import new data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(['Luffy', 'Zorro', 'Nami', 'Usopp', 'Sanji'])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# let's look at subplots - two figures next to each other
# compare relative positions
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, 
    nrows=1,
    figsize=(7,3)) # creates two plots, using one row, but each graphs gest its own columns
ax1.scatter(x1, y1)
ax2.bar(x2, y2)
```

```python
# make each subplot different

fig, (ax1, ax2) = plt.subplots(ncols=2, 
    nrows=1,
    figsize=(7,3)) # creates two plots, using one row, but each graphs gest its own columns
ax1.scatter(x1, y1, marker = '+')
ax2.bar(x2, y2, color='orange')

# modify ax1 - label x and y axis
ax1.set_xlabel('Episode Number')
ax1.set_ylabel('Duration of episode (min)')
ax1.set_title('One Piece Episode Duration')

# modify ax2 - label x and y axis
ax2.set_xlabel('Character')
ax2.set_ylabel('Number of Appearances per Episode')
ax2.yaxis.set_label_position('right')
ax2.yaxis.tick_right()
ax2.set_title('One Piece Characters')
```

##### Grid Arrangements #####

```python
# without grid arrangement?
# using someaxes - group name for a bunch of axes

# arragenment of axes is how the plots will be displayed
# creates a mosaic, the position of ax1, ax2, ax3 is where they will be located
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                   ['ax2', 'ax3']],
                                   figsize=(7,4))

someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1,y1)
```

```python
# see the mosaic in our data
# ax1 is top left, ax2 is the right side, ax3 is the bottom left
```

##### Modify Figure Layout #####

```python
# adding x axis labels
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                   ['ax2', 'ax3']],
                                   figsize=(7,4), 
                                   layout='constrained')

someaxes['ax1'].scatter(x1, y1)
someaxes['ax2'].bar(x2, y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel("A big label", fontsize=18)
someaxes['ax2'].set_xlabel('Another label', fontsize=18)
someaxes['ax3'].set_xlabel('Label2: 2 Fast 2 Furious', fontsize=18)
# data is cut off somewhat
# but try adding , layout='constrained' within the mosaic command
# the data gets 'compressed' so its all viewable in the mosaic
```
