```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import PIL
import requests
```

```python
# load the data
# using tips data
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
sns.set_style('whitegrid')
# can use darkgrid, or other options
sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
# same thing with changing axis names and titles
tipgraph = sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')

tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
# see how the grid is set up now on our plots
```

```python
# setting size of figure + elements of the lineplot
fig = plt.subplots(figsize=(10,3))

tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')
tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
# see the edits with line, colour, marker etc.
```

```python
# very useful in adding multiple variables for different visual elements of the graph
# change colour based on values

tipgraph = sns.scatterplot(data = tips,
                           x = 'total_bill',
                           y = 'tip',
                           style = 'time', # shape for type of meal
                           hue = 'day', # colour for each day
                           palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])

tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')

# see how colour/shape changes dependent on the day and meal
```

```python
# pairplot - multi-dimensional comparisons of variables
# for the datasets we are looking at

# data and the variables we are looking at
sns.pairplot(data = tips,
            hue = 'day')
# easy way to look at all these relationships
# see comparison of day across the different variables
```

```python
# rel plot uses some specific variables
daysplot = sns.relplot(data = tips,
                       x = 'total_bill', # variable along x axis
                       y = 'tip', # variables along y axis
                       hue = 'sex',  # colour of points broken into groups of this variable
                       col = 'day', # each column is one day
                       kind = 'scatter', # the graphs are scatter plots
                       col_wrap = 2 # we have two days per column
)
# remove the col wrap - each day will have its own column
```

```python
# plotly
import plotly.graph_objects as go 

x1 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
# plotly is different
# use go.Figure()
# add_trace(go.Bar) -> this creates our graph objects
# similar to before, we specify elements on the graph

graph = go.Figure()

graph.add_trace(go.Bar(x = x1, y=y1))
graph.update_layout(title = 'Pirate Scores',
                    xaxis_title = 'Pirates',
                    yaxis_title = 'Score')
# see the interaction of what happens
# we can modify the other elements soon
# can download as well, highlight certain areas, see top right corner
# similar elements to matplotlib but not entirely the same
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(
    x = x1,
    y = y1,
    mode = 'markers',
    marker = dict(
        size = 15,
        color = 'hotpink',
        opacity = 1,
        line = dict(width = 5,
                   color = 'purple')
    )
))

graph.update_layout(title = 'Interactive Pirates',
                    xaxis_title = 'Pirates',
                    yaxis_title = 'Scores',
                    width = 500,
                    height = 500)
# see the change to the graph and the new interactivity
# as a reminder, each library has its own documentation that can describe what happens with the code
```

```python
# Word clouds and venn diagrams

# could always use ms paint or photoshop
# but remember reproducible data here

from wordcloud import WordCloud

df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv',
                 on_bad_lines = 'skip')
df
```

```python
# text to join data from the quote column
text = " ".join(x for x in df.quote)
text
# see how we get all the text combined together
```

```python
wordcloud = WordCloud(background_color= 'white',
                      colormap = 'inferno').generate(text) 
# this creates a word cloud object that we can plot against the axes we describe in mpl
# bilinear makes image smoother and fit together nicer

fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud, 
          interpolation = 'bilinear')
ax.axis('off')
```

```python
# venn diagrams
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(['apple', 'banana', 'watermelon'])
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])
```

```python
# make the ven diagrams

diagram = venn2([A,B],
                set_labels = ('Fruits', 'Pie'),
                set_colors = ('blue', 'red'),
                alpha = 0.5)

# we want to see the actual lists of words, rather than count in each circle
```

```python
# include the words
# https://github.com/konstantint/matplotlib-venn/blob/master/matplotlib_venn/_common.py#L12

diagram = venn2([A,B],
                set_labels = ('Fruits', 'Pie'),
                set_colors = ('blue', 'red'),
                alpha = 0.5) # important value to help reduce intensity of colour since its opacity
diagram.get_label_by_id('10').set_text("\n".join(A-B)) # separates each text item by a new line (\n)
# done only for A set

diagram.get_label_by_id('11').set_text("\n".join(A&B)) # A and B
diagram.get_label_by_id('01').set_text("\n".join(B-A)) # set B and not A
# the numbers are the positioning of the text
# can find the explanation in the docs
# see the interaction of A/B and A and B
```
